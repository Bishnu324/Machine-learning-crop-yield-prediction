```python
import pandas as pd
from mlxtend.feature_selection import SequentialFeatureSelector
```

```python
data=pd.read_csv("C:\\Users\\A S P I R E 7\\OneDrive - mishu\\Desktop\\python course\\diabetes.csv")
```

```python
data.head(5)
```

```python
x=data.iloc[:,:-1]
y=data["Outcome"]
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
fs=SequentialFeatureSelector(lr,k_features=6,forward=True)
fs.fit(x,y)
```

```python
fs.feature_names
```

```python
fs.k_feature_names_
```

```python
fs.k_score_
```
